```python
%load_ext autoreload
%autoreload 2
import spikeinterface.full as si
```

```python
rec, sorting = si.generate_ground_truth_recording(num_channels=128, num_units=250)
sa = si.create_sorting_analyzer(sorting, rec)
```

```python
si.set_global_job_kwargs(n_jobs=0.8)
sa.compute(['random_spikes', 'templates'])
```

```python
res = {}
for method in ['union', 'intersection', 'dense']:
    print(method)
    res[method] = sa.compute('template_similarity', support=method, method='l1').get_data()
import pylab as plt
fig, axes = plt.subplots(2, len(res.keys()), figsize=(15, 5))
for count, key in enumerate(res.keys()):
    axes[0, count].imshow(res[key])
    axes[0, count].set_title(key)
    axes[1, count].hist(res[key].flatten(), 100)
    axes[1, count].set_yscale('log')
```
